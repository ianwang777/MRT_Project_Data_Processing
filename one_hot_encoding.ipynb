```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
holiday = pd.read_csv('HOLIDAY_data/holiday.csv')
covid = pd.read_csv('COVID_cooked_data/COVID_cases_per_day_policy.csv')
```

##### holiday one hot

```python
holiday_onehot = pd.get_dummies(holiday.iloc[:,1])
holiday_onehot = pd.concat( [ holiday['日期'], holiday_onehot ], axis=1)
holiday_onehot.to_csv(f"HOLIDAY_data/holiday_onehot.csv", index=False)
holiday_onehot
```

##### covid one hot

```python
covid.columns
```

```python
new_mode = pd.get_dummies(covid['經濟防疫新模式'])
new_mode.columns = ['無經濟防疫新模式','有經濟防疫新模式']

second_alert = pd.get_dummies(covid['二級警戒'])
second_alert.columns = ['無二級警戒','有二級警戒']

third_alert = pd.get_dummies(covid['三級警戒'])
third_alert.columns = ['無三級警戒','有三級警戒']

seven_seven = pd.get_dummies(covid['是否7+7'])
seven_seven.columns = ['無7+7','有7+7']

three_four = pd.get_dummies(covid['是否3+4'])
three_four.columns = ['無3+4','有3+4']

first_case = pd.get_dummies(covid['首例確診'])
first_case.columns = ['無首例確診','有首例確診']

mask_policy = pd.get_dummies(covid['口罩實名制'])
mask_policy.columns = ['無口罩實名制','有口罩實名制']

sms_report = pd.get_dummies(covid['簡訊實聯制上路'])
sms_report.columns = ['無簡訊實聯制','有簡訊實聯制']

delta_mutate = pd.get_dummies(covid['Delta變異株'])
delta_mutate.columns = ['無Delta變異株','有Delta變異株']

omicron_mutate = pd.get_dummies(covid['Omicron亞型變異株'])
omicron_mutate.columns = ['無Omicron亞型變異株','有Omicron亞型變異株']

list_to_concat = [covid['日期'], covid['累積確診人數'], covid['確診人數'], new_mode, second_alert, third_alert, seven_seven, three_four, 
                 first_case, mask_policy, sms_report, delta_mutate, omicron_mutate]
covid_onehot = pd.concat( list_to_concat, axis=1)
covid_onehot.to_csv(f'COVID_cooked_data/covid_onehot.csv', index=False)
covid_onehot
```

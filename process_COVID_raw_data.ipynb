```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('COVID_raw_data/Day_Confirmation_Age_County_Gender_19CoV.csv')
df.replace('/','-',inplace=True,regex=True)
df
```

```python
df.groupby(['個案研判日']).sum()
df.groupby(['個案研判日']).sum().shape
```

```python
date = pd.DataFrame(  df.groupby(['個案研判日']).sum().index  )
case = pd.DataFrame(  df.groupby(['個案研判日']).sum().values  )
cases_per_day = pd.concat([date,case],axis=1)

# cases_per_day.shape
cases_per_day.columns = ['日期','確診人數']
cases_per_day
```

```python
if not os.path.isdir('COVID_cooked_data'):
    os.mkdir('COVID_cooked_data')
cases_per_day.to_csv(f"COVID_cooked_data/COVID_cases_per_day.csv",index=False)
```
